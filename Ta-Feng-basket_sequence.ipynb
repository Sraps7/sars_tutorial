```python
df['date_time'] = df['date_time'].apply(lambda x: datetime.strptime(x[:10], '%m/%d/%Y'))
```

```python
df['trans_id'] =  df['customer_id'].astype(str) + df['date_time'].astype(str)
```

```python
df = df.sort_values(['trans_id'])
```

```python
cust_id = df.customer_id.values[0]
trans_buf_id = df.trans_id.values[0]
trans = 0
trans_id_list = []
```

```python
for i in tqdm(range(df.shape[0])):
    if df.customer_id.values[i] == cust_id:
        if df.trans_id.values[i] == trans_buf_id:
            trans_id_list.append(trans)
        else:
            trans += 1
            trans_buf_id = df.trans_id.values[i]
            trans_id_list.append(trans)
    else:
        cust_id = df.customer_id.values[i]
        trans_buf_id = df.trans_id.values[i]
        trans = 0
        trans_id_list.append(trans)
df['trans_id'] = trans_id_list
```

```python
def create_sequence_for_one_customer(df_cust: pd.DataFrame):
    #if df_cust.trans_id.max() < 4:
        #return
    basket_seq = ''
    features = []
#df_cust_1 = df.iloc[start:10,:].copy()
    for i in range(df_cust.trans_id.max() + 1):
        curr_product = set(df_cust.loc[df_cust.trans_id == i,'product_id'])
        features.append(curr_product)
    #print(features)
    for j in range(len(features)):
        if j<len(features)-1:
            basket = ''
            list(features[j])
            for l in range(len(list(features[j]))):
                #print(type(item))
                if l<len(list(features[j]))-1:
                    str_item = str(list(features[j])[l])
                    basket = basket+str_item+' '
                else:
                    str_item = str(list(features[j])[l])
                    basket = basket+str_item
            basket_seq = basket_seq+basket+'|'
        else:
            basket = ''
            list(features[j])
            for l in range(len(list(features[j]))):
                #print(type(item))
                if l<len(list(features[j]))-1:
                    str_item = str(list(features[j])[l])
                    basket = basket+str_item+' '
                else:
                    str_item = str(list(features[j])[l])
                    basket = basket+str_item
            basket_seq = basket_seq+basket  
    return basket_seq
           #print(basket_seq)
    
```
